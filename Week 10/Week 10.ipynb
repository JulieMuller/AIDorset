```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
#split the dataset in training and testing sets
X, y = make_circles(n_samples=10000, noise=0.05, random_state=26)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=33, random_state=26)

#Graphs of the data
fig, (train_ax, test_ax) = plt.subplots(ncols = 2, sharex = True, figsize = (10, 5))

#training data
train_ax.scatter(X_train[:,0], X_train[:,1], c = y_train)
train_ax.set_title("Training data")
train_ax.set_xlabel("X_train")
train_ax.set_ylabel("y_train")

#test data
test_ax.scatter(X_test[:,0], X_test[:,1], c = y_test)
test_ax.set_title("Testing data")
test_ax.set_xlabel("X_test")
test_ax.set_ylabel("y_test")
```

```python
from torch import nn #neural network
from torch import optim #gradient descent optimization
```

```python
input_dim = 2 #first layer 2 neurons
hidden_dim = 2 #10 neurons 
output_dim = 1 #one output
```

```python
class NeuralNetwork (nn.Module): 

    def __init__(self, input_dim, hidden_dim, output_dim):
        super(NeuralNetwork, self).__init__() 
        self.layer_1 = nn. Linear(input_dim, hidden_dim)
        nn.init.kaiming_uniform (self.layer_1.weight, nonlinearity="relu")
        self.layer_2 = nn. Linear (hidden_dim, output_dim)

    def forward(self, x):
        x = torch.nn.functional.relu (self.layer_1(x))
        x = torch.nn.functional.sigmoid (self.layer_2(x))
        return x
```
